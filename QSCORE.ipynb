### Signal

Instead of Z-Score we use an unbiased Q-score defined as follows

$$Q[t] = \frac{S[t] - \tau_{50}\left(S[t]\right)}{\tau_{75}\left(S[t]\right) - \tau_{25}\left(S[t]\right)}$$

where $S[t]$ is the value of the signal for a given pair at time $t$.

### Position

Instead of a binary position (+/-) we scale with the magnitude of the z-score. Specifically:

$\newcommand{\sgn}{\operatorname{sgn}}$

$\newcommand\round[1]{\left[#1\right]}$

$$P[t] = \sgn{(Q[t])}\round{|Q[t]|}$$

where $[]$ indicates rounding. We only use the last 20 days (~1 trading month) to calculate today's score and apply a shift.
